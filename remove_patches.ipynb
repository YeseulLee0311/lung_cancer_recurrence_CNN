```python
import cv2
import matplotlib.pyplot as plt
import os
import shutil
```

```python
path='./TCIA_new'
```

```python
if not(os.path.isdir(os.path.join(path,'peritumoral_patches_removed0.1'))):
    os.makedirs(os.path.join(path,'peritumoral_patches_removed0.1'))
if not(os.path.isdir(os.path.join(path,'peritumoral_patches_removed0.15'))):
    os.makedirs(os.path.join(path,'peritumoral_patches_removed0.15'))
```

```python
###threshold ==0.1
for file_name in os.listdir(os.path.join(path,'peritumoral_patches')):
    patch=cv2.imread(os.path.join(path,'peritumoral_patches',file_name),cv2.IMREAD_GRAYSCALE)
    count=0
    for x in range(len(patch)):
        for y in range(len(patch[x])):
            if patch[x][y]!=0:
                count+=1
    if count/(patch.shape[0]*patch.shape[1])>=0.1:
        shutil.copy(os.path.join(path,'peritumoral_patches',file_name),os.path.join(path,'peritumoral_patches_removed0.1'))
    else:
        print("removed patch: ", file_name)
        plt.imshow(patch,cmap='gray')
        plt.show()
```
