```python
import pandas as pd
import os
```

```python
#load csv file and select columns
org=pd.read_csv('./NSCLC Radiomics Lung1.clinical-version3-Oct 2019.csv',sep=',')
org=org[['PatientID','Survival.time']]
```

```python
#our patient list to DataFrame
new=pd.DataFrame(os.listdir('./TCIA_test'),columns=['PatientID'])
```

```python
#merge org with new , key='PatientID'
new=pd.merge(org,new,on='PatientID')
```

```python
new
```

```python
#convert to month unit and check for recurrence
survival_month=[]
recurrence=[]
for i in range(len(new)):
    survival_month.append(int(new['Survival.time'][i]/30))
    if survival_month[i]<=24:
        recurrence.append(1)
    else:
        recurrence.append(0)
```

```python
#add new columns
new['Survival.time(month)']=survival_month
new['recurrence']=recurrence
```

```python
new
```

```python
#'new' DataFrame to csv
new.to_csv('./label.csv',index=False)
```
